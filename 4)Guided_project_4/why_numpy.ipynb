```python
revenues = [2000,5000,3000,6500,7200,3100,2650,1900,3820,5120,4100]
```

```python
import time
```

```python
initial_time = time.time()
sum = 0
```

```python
for i in revenues:
  sum += i
print(sum)
termination_time = time.time()
 
print("Execution Time: ",termination_time-initial_time)
```

```python
array = np.array(revenues)
 
initial_time = time.time()
 
sum = array.sum()
print(sum)
termination_time = time.time()
 
print("Execution Time: ",termination_time-initial_time)
```
